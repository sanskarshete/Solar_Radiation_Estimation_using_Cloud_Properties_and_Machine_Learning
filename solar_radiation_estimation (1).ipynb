```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
X_train=data.iloc[:750,:-1]
X_test=data.iloc[750:,:-1]
y_train=data.iloc[:750,-1]
y_test=data.iloc[750:,-1]
```

```python
# X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=20)
```

```python
scaler=MinMaxScaler().fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
print(X_train.shape,"is x-train shape")
print(X_test.shape,"is x-test shape")
print(y_train.shape,"is y-train shape")
print(y_test.shape,"is y-test shape")
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
```

```python
model=Sequential()
model.add(Dense(128,activation='elu'))
model.add(BatchNormalization())
model.add(Dense(100,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(Dense(32,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1))

model.compile(loss='mse',optimizer='SGD',metrics=['mse'])
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=150,batch_size=70,steps_per_epoch=11)
```

```python
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
predict=model.predict(scaler.transform(x))
```

```python
predict.shape
```

```python
import math
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_percentage_error,mean_absolute_error


```

```python
def index_of_agreement(obs, pred):
    mean_obs = sum(obs) / len(obs)
    numerator = sum(abs(obs - pred))
    denominator = sum(abs(obs - mean_obs)) + sum(abs(pred - mean_obs))
    return 1 - numerator / denominator
```

```python
training_rmse=math.sqrt(mean_squared_error(y_train,train_predict))
testing_rmse=math.sqrt(mean_squared_error(y_test,test_predict))
testing_mae=mean_absolute_error(y_test,test_predict)
testing_mape=mean_absolute_percentage_error(y_test,test_predict)
kappa_score= index_of_agreement(y.ravel(),predict.ravel())
r2=r2_score(y,predict)
```

```python
print("TRAINING RMSE OF ANN ",training_rmse)
print("TESTING RMSE OF ANN ",testing_rmse)
print("MEAN ABSOLUTE ERROR OF ANN ",testing_mae)
print("TESTING MAPE OF ANN ",testing_mape)
print("NDEX OF AGREEMENT OF ANN ",kappa_score)
print("R2 SCORE OF ANN ",r2)
```
